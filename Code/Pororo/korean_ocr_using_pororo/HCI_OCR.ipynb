# Pororo OCR

**Pororo란?** Pororo는 Platform Of neuRal mOdels for natuRal language prOcessing의 약어이며,카카오브레인에서 다양한 한글 자연어 처리를 위해 통합된 형태의 자연어 프레임워크입니다. Pororo가 다양한 기능을 제공하지만 그 중 필요한 OCR(Optical Character Recognition) 기능만을 추출하여 화장품 성분 글자인식을 시도하였습니다. 

**분석방법:** 사용자는 자신에게 맞지 않다고 생각하는 화장품의 성분 사진을 어플 내에 저장할 것입니다. 저희는 OCR 기술을 활용하여 글자인식을 진행하여 공통성분을 뽑아 낸 뒤, 공통성분이 1) 피부타입별 위험성분, 2) 알레르기 유발 주의 성분, 3) 20가지 주의성분 중에 포함되면 사용자가 조심해야 하는 성분으로 분류하였습니다. (이 프로젝트에서는 2가지의 화장품으로만 공통성분을 선별하였지만, 실제로 어플을 사용하면 2개 이상의 화장품을 등록하는 것도 가능합니다.)
