## User Information

```python
file = open(r'./user_info.txt')
user_info=file.readlines()

name=user_info[0]
name=name.strip()
print("--- "+str(name)+"님의 정보 ---")

age=user_info[1]
age = age.strip()
print('age: ',age)

skinType=user_info[2]
skinType=skinType.strip()
print('skinType: ',skinType)

skinConcern=user_info[3]
print('skinConcern: ',skinConcern)
```

## Import libraries

```python
import pandas as pd
import re
```

## Dataset Upload

```python
Men=pd.read_csv('./men.csv')
pd.set_option('display.max_rows',None)
```

```python
Ranked = Men[Men['나이'].str.contains('20대,30대')]
print(len(Ranked))
```

```python
Ranked = Men[Men['나이'].str.contains(age)]
Ranked = Ranked[Ranked['피부타입'].str.contains(skinType)]
Ranked = Ranked[Ranked['피부고민'].str.contains(skinConcern)]

#If considered 2 above, but if length is 0: consider only skinConcern
if len(Ranked)==0:
    Ranked = Men[Men['피부고민'].str.contains(skinConcern)]
    
# others
others=pd.merge(Ranked, Men, how='outer',indicator=True)
others=others.query('_merge=="right_only"').drop(columns=['_merge'])

#reset index
Ranked.reset_index(inplace=True)
Ranked = Ranked.drop(['index', '랭킹','리뷰수'], axis=1)
Ranked.index=Ranked.index + 1
others.reset_index(inplace=True)
others = others.drop(['index', '랭킹','리뷰수'], axis=1)
others.index=others.index + 1
```

## OCR

```python
ocr_result=[]

with open("./ocr.txt", "r") as ocr_file:
    for line in ocr_file:
        line=line.strip()
        ocr_result.append(line)

print("OCR Result: ",ocr_result)
```

## Recommendation Ranking after OCR

```python
OCR_Ranked = Ranked
for i in range (len(ocr_result)):
    OCR_Ranked = OCR_Ranked[OCR_Ranked['성분'].str.contains(ocr_result[i])==False]
OCR_Ranked.reset_index(inplace=True)
OCR_Ranked.index=OCR_Ranked.index + 1
print("제외성분갯수:",len(Ranked)-len(OCR_Ranked))
```

```python
OCR_Ranked = OCR_Ranked.drop(['index'], axis=1)
OCR_Ranked
```

## Others Ranking after OCR

```python
OCR_others = others
for i in range (len(ocr_result)):
    OCR_others = OCR_others[OCR_others['성분'].str.contains(ocr_result[i])==False]
OCR_others.reset_index(inplace=True)
OCR_others.index=OCR_others.index + 1
print("제외성분갯수:",len(others)-len(OCR_others))
```
