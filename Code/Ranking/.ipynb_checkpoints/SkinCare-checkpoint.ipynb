## User Information

```python
file = open(r'./user_info.txt')
user_info=file.readlines()

name=user_info[0]
name=name.strip()
print("--- "+str(name)+"님의 정보 ---")

age=user_info[1]
age = age.strip()
print('age: ',age)

skinType=user_info[2]
skinType=skinType.strip()
print('skinType: ',skinType)

skinConcern=user_info[3]
print('skinConcern: ',skinConcern)
```

## Import libraries

```python
import pandas as pd
import re
```

## Dataset Upload

```python
SkinCare=pd.read_csv('./skinCare.csv',encoding='cp949')
pd.set_option('display.max_rows',None)
```

```python
Ranked_SkinCare = SkinCare[SkinCare['나이'].str.contains(age)]
Ranked_SkinCare = Ranked_SkinCare[Ranked_SkinCare['피부타입'].str.contains(skinType)]
Ranked_SkinCare = Ranked_SkinCare[Ranked_SkinCare['피부고민'].str.contains(skinConcern)]

#If considered 3 above, but if length is 0: consider only skinConcern
if len(Ranked_SkinCare)==0:
    Ranked_SkinCare = SkinCare[SkinCare['피부고민'].str.contains(skinConcern)]
    
# others
others=pd.merge(Ranked_SkinCare, SkinCare, how='outer',indicator=True)
others=others.query('_merge=="right_only"').drop(columns=['_merge'])

#reset index
Ranked_SkinCare.reset_index(inplace=True)
Ranked_SkinCare = Ranked_SkinCare.drop(['index', '랭킹','리뷰수'], axis=1)
Ranked_SkinCare.index=Ranked_SkinCare.index + 1
others.reset_index(inplace=True)
others = others.drop(['index', '랭킹','리뷰수'], axis=1)
others.index=others.index + 1
```

## Recommendation Ranking
나이대, 피부타입, 피부고민에 의해 선정된 랭킹 (만약 데이터가 적다면 오직 피부타입으로만 결정됨)

```python
Ranked_SkinCare
```

## Others Ranking
Recommendation Ranking에 제시된 화장품을 제외한 랭킹

```python
others
```

## OCR

```python
ocr_result=[]

with open("./ocr.txt", "r") as ocr_file:
    for line in ocr_file:
        line=line.strip()
        ocr_result.append(line)

print("OCR Result: ",ocr_result)
```

## Recommendation Ranking after OCR

```python
OCR_Ranked_SkinCare = Ranked_SkinCare
for i in range (len(ocr_result)):
    OCR_Ranked_SkinCare = OCR_Ranked_SkinCare[OCR_Ranked_SkinCare['성분'].str.contains(ocr_result[i])==False]
OCR_Ranked_SkinCare.reset_index(inplace=True)
OCR_Ranked_SkinCare.index=OCR_Ranked_SkinCare.index + 1
print("제외성분갯수:",len(Ranked_SkinCare)-len(OCR_Ranked_SkinCare))
```

```python
OCR_Ranked_SkinCare = OCR_Ranked_SkinCare.drop(['index'], axis=1)
OCR_Ranked_SkinCare
```

## Others Ranking after OCR

```python
OCR_others = others
for i in range (len(ocr_result)):
    OCR_others = OCR_others[OCR_others['성분'].str.contains(ocr_result[i])==False]
OCR_others.reset_index(inplace=True)
OCR_others.index=OCR_others.index + 1
print("제외성분갯수:",len(others)-len(OCR_others))
```

```python
OCR_others = OCR_others.drop(['index'], axis=1)
OCR_others
```
